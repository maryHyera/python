```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import *
import pickle

%matplotlib inline
```

```python
df = pd.read_csv("Handwritting.csv")
```

```python
df.head()
```

```python
df.dropna(axis=1, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn import preprocessing
```

```python
df = pd.concat([df, pd.get_dummies(df['Pressure'], prefix='pressure')],axis=1)
df.drop(['Pressure'],axis=1, inplace=True)
```

```python
df.head()
```

```python
df = pd.concat([df, pd.get_dummies(df['Strings'], prefix='strings')],axis=1)
df = pd.concat([df, pd.get_dummies(df['slant'], prefix='slant')],axis=1)
```

```python
df.drop(['Strings'],axis=1, inplace=True)
df.drop(['slant'],axis=1, inplace=True)
```

```python
df.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
label_encoder.fit(df["Size"])
```

```python
df["size"] = label_encoder.transform(df["Size"])
df.drop(['Size'],axis=1, inplace=True)
df
```

```python
label_encoder.fit(df["Spacing"])
df["Spacing"] = label_encoder.transform(df["Spacing"])
df.tail()
```

```python
label_encoder.fit(df["Gaps"])
df["Gaps"] = label_encoder.transform(df["Gaps"])
df.head()
```

```python
label_encoder.fit(df["Label"])
df["Label"] = label_encoder.transform(df["Label"])
df.head()
```

```python
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df[['Spacing','Gaps','pressure_Average','pressure_Hard','pressure_Light','slant_Left','slant_Right','slant_straight','size']]
y = df['Label']
```

```python
kn = KNeighborsClassifier()
kn.fit(X,y)
```

### Trying with SVM model


```python
#splitting the data into train and test set
from sklearn.model_selection import train_test_split
import sklearn
print('version is {}'.format(sklearn.__version__))
```

```python
X = df[['Spacing','Gaps','pressure_Average','pressure_Hard','pressure_Light','slant_Left','slant_Right','slant_straight','size']]
y = df['Label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
#importing svm mode
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

### Grid search

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]} 
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid.fit(X_train,y_train)
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
