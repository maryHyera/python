```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline     #This allows you to see things inside the jupyter notebook
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

### functional way of ploting graphs

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('TITLE')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object oriented

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('y label')
axes.set_title('TITLE')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.2,1,1])
axes2 = fig.add_axes([0.2,0.65,0.5,0.5])

axes1.plot(x,y)
axes1.set_title('LARGE PLOT')
axes2.plot(y,x)
axes2.set_title('SMALL PLOT')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
plt.tight_layout()
```
